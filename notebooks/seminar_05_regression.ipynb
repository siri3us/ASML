# Обусловленность и регуляризация [[toc]](#toc)

Рассмотрим некоторый сигнал $f \in \mathbb{R}^n$. Мы не знаем его реальное значение, т.к. воспринимаем его с помощью некоторого прибора. Прибором в нашей терминологии является линейный оператор $A: \mathbb{R}^n \to \mathbb{R}^n$, влияющий на истинный сигнал $f$ и преобразующий его в воспринимаемое нами предтавление.

Т.е. мы видим результат измерений $\xi = Af + \nu$

```python
amplitudes  = np.array([10.0, 1.0])
frequencies = np.array([1. / 32, 1. / 2])
periods = 1.0 / frequencies
angle_frequencies = 2 * np.pi * frequencies
# Signal f dimension.
n_points = 256

time_start = 0
time_end = np.max(periods)
t = np.linspace(time_start, time_end, n_points).reshape(-1, 1)
time_step = (time_end - time_start) / (n_points - 1)

f = np.sum(amplitudes * np.sin(t * angle_frequencies), axis=1)

plt.plot(t, f, linewidth=2, color='b')
plt.grid(which='both', linestyle='--')
plt.xlabel('x')
tmp_ = plt.ylabel('sum of sines')
```

Вот так выглядит оператор $A$:

```python
A = np.eye(n_points)
filter_size = 10
filter_span = filter_size // 2
for i in range(n_points):
    A[i, max(i - filter_span, 0): min(i + filter_span + 1, n_points)] = 1
A /= filter_size
```

```python
plt.imshow(A)
tmp_ = plt.colorbar()
```

Запишем результат измерения сигнала прибором:

```python
sigma = 0.1
nu = np.random.normal(0, sigma, f.shape)
xi = A.dot(f) + nu

plt.plot(xi, label='noisy')
plt.plot(f, label='clean')
plt.xlabel('x')
plt.ylabel('signal value')
tmp_ = plt.legend()
```

### Оценка наименьших квадратов

Хотелось бы по $\xi$ построить оценку реального сигнала -- т.е. решить задачу регрессии.

MLE оценки в случае нормального шума получены на лекции / семинаре / самостоятельно:
$$
\xi = Af + \nu, \quad \nu \sim \mathcal{N}(0, \sigma^2)\\
\hat f = \underbrace{(A^TA)^{-1}A^T}_{R} \cdot \xi
$$

```python
R = np.linalg.inv(A.T.dot(A)).dot(A.T)
f_estimate = R.dot(xi)
```

```python
plt.plot(xi, label='noisy')
plt.plot(f, label='clean')
plt.plot(f_estimate, label='estimate')
plt.xlabel('x')
plt.ylabel('signal value')
plt.title('Oops...')
tmp_ = plt.legend()
```

Что случилось? Случилось то, что матрица плохо обусловлена и всё сломалось.

Можно напрямую вычислить псевдообратную матрицу для $A$: $A^{+} = (A^TA)^{-1}A^T$.

```python
R = np.linalg.pinv(A)
f_estimate = R.dot(xi)
```

```python
plt.plot(xi, label='noisy')
plt.plot(f, label='clean')
plt.plot(f_estimate, label='estimate')
plt.xlabel('x')
plt.ylabel('signal value')
tmp_ = plt.legend()
```

На самом деле, понять, что всё будет плохо, можно было просто взглянув на [число обусловленности](http://www.machinelearning.ru/wiki/index.php?title=%D0%9C%D0%BD%D0%BE%D0%B3%D0%BE%D0%BC%D0%B5%D1%80%D0%BD%D0%B0%D1%8F_%D0%BB%D0%B8%D0%BD%D0%B5%D0%B9%D0%BD%D0%B0%D1%8F_%D1%80%D0%B5%D0%B3%D1%80%D0%B5%D1%81%D1%81%D0%B8%D1%8F#.D0.9F.D1.80.D0.BE.D0.B1.D0.BB.D0.B5.D0.BC.D1.8B) $A$ -- отношение максимального и минимального СЗ:

```python
np.linalg.cond(A)
```

### Добавление $l_2$ регуляризации
Попробуем перед обращением матрицы добавить к её диагонали небольшие числа:
$$
\tilde R = (A^TA + \lambda I)^{-1}A^T
$$
Тем самым мы надеемся получить невырожденную матрицу, практически не повлияв на её значения.

```python
# Change that for some more appropriate value.
reg_coeff = 0.001
R = np.linalg.inv(A.T.dot(A) + reg_coeff * np.eye(A.shape[0])).dot(A.T)
f_estimate = R.dot(xi)
```

```python
plt.plot(xi, label='noisy')
plt.plot(f, label='clean')
plt.plot(f_estimate, label='estimate')
plt.xlabel('x')
plt.ylabel('signal value')
tmp_ = plt.legend()
```

## Модель линейной регрессии<sup>[toc](#toc)</sup>
Предполагается, что наблюдаемые значения вектора $\boldx \in \RR^d$ (зависимых переменных) и $y \in \RR$ (независимой переменной) связаны следующим образом
$$
y = \boldx^T\boldw + \eps, \quad \eps \sim \Normal(0, \sigma^2)
$$
Или в матричном виде
$$
\boldy = \boldX\boldw + \boldeps, \quad \boldeps \sim \Normal(\boldzero, \sigma^2 \boldI)
$$

Сумма квадратов невязок (**residual sum of squares**):
$$
\RSS(\boldw) = \|\boldX\boldw - \boldy\|_2^2
$$

В качестве оценки весов $\boldw$ выбираются веса, минимизирующие значения $\RSS(\boldw)$:
$$
\RSS(\boldw) \rightarrow \min_{\boldw}
$$
Полученая оценка весов $\hatboldw$ называется OLS-оценкой (Ordinary Least Squares). На предыдущем семинаре для этой оценки получили следующие свойства:
\begin{align*}
&\hatboldw = (\boldX^T\boldX)^{-1}\boldX^T\boldy\\
&\Exp_{\boldw} \hatboldw = \boldw, \quad \Cov_{\boldw}(\hatboldw) = \sigma^2(\boldX^T\boldX)^{-1}\\
&\hatsigma^2 = \frac{1}{n - d}\RSS(\hatboldw),\quad\Exp_{\boldw}\ls\RSS(\hatboldw)\rs = (n - d)\sigma^2, \quad \Exp_{\boldw}\hatsigma^2 = \sigma^2
\end{align*}

### Вероятностная интерпретация линейной регрессии<sup>[toc](#toc)</sup>

$$
\boldy = \boldX\boldw + \boldeps,\quad \boldy\in\RR^n, \quad \boldX \in \RR^{n\times d}, \quad \boldeps \sim \Normal(\boldzero,\sigma^2\boldI_n)
$$

В качестве неизвестных параметров выступают $\boldw$ и $\sigma^2$. 

Запишем правдоподобие выборки:
$$
\mathcal{L}(\boldw,\sigma^2;\boldX,\boldy) = \Sum_{i=1}^n \log p(y_i|\boldx_i) = \Sum_{i=1}^n \lf -\frac{1}{2} \log(2\pi) - \frac{1}{2}\log \sigma^2 - \frac{1}{2\sigma^2} (\boldw^T\boldx_i - y_i)^2\rf = \text{const} - \frac{n}{2} \log \sigma^2 - \frac{1}{2\sigma^2} \|\boldX\boldw - \boldy\|_2^2
$$
Максимизация правдоподобия по $\boldw$ соответствует минимизации $RSS(\boldw)$. А это задачу мы уже решили
$$
\mleboldw = (\boldX^T\boldX)^{-1}\boldX^T\boldy
$$
Находя максимум правдоподобия по $\sigma^2$, получим
$$
\mlesigma^2 = \frac{1}{n}\|\boldX\boldw - \boldy\|_2^2 = \frac{1}{n}\RSS(\mleboldw)
$$

### Простая линейная регрессия<sup>[toc](#toc)</sup>
$$
y = w_0  + w_1 x + \eps
$$

$$
\boldX = \begin{pmatrix}
1 & x_1\\
1 & x_2\\
\vdots &\vdots \\
1 &x_N
\end{pmatrix},\quad
\boldw = \begin{pmatrix}
w_0\\
w_1
\end{pmatrix}
$$

$$
\Cov \begin{pmatrix}
\hatw_0\\
\hatw_1
\end{pmatrix} =
\sigma^2 (\boldX^T\boldX)^{-1} =
\begin{pmatrix}
\sum_i x_i^2 &\sum_i x_i\\
\sum_i x_i   &n
\end{pmatrix}^{-1} = 
 \sigma^2
\begin{pmatrix}
n \mean{\boldx^2} & n\mean{\boldx}\\
n \mean{\boldx}   & n
\end{pmatrix}^{-1} =
\frac{\sigma^2}{n\lp\mean{\boldx^2} - \mean{\boldx}^2\rp}
\begin{pmatrix}
\mean{\boldx^2} & -\mean{\boldx}\\
-\mean{\boldx}  & 1
\end{pmatrix}
$$

### p-value значений весов<sup>[toc](#toc)</sup>

Если $\boldeps \sim \Normal(\boldzero, \sigma^2\boldI)$, то
$$
\hatboldw \sim \Normal(\underbrace{(\boldX^T\boldX)^{-1}\boldX^T\boldy}_{\boldmu}, \underbrace{\sigma^2(\boldX^T\boldX)^{-1}}_{\Sigma})
$$

$$
\boldw = (w_1, \dots, w_n)^T \in \RR^n
$$

$$
p(\hatw_i) = \Normal(\hatw_i; w_i, \sigma_i^2), \quad \sigma_i^2 = \Covariance_{ii}
$$

Можем проверить гипотезу $H_0\colon \hatw_i = 0$ против альтернативы $H_0\colon \hatw_i \neq 0$. Для этого воспользуемся критерием Вальда:
$$
W = \frac{\hatw_i}{\sigma_i}, \qquad \pvalue = 2\Phi(-|W|)
$$

### Чувствительность оценки весов по отношение к независимым переменным $\boldX$<sup>[toc](#toc)</sup>

Пусть $\boldX \in \RR^{n\times d}$ &mdash; матрица объектов-признаков причем $n \ge d$. Пусть дано ее SVD-разложение
$$
\boldX = UDV^T
$$
где $U \in \RR^{n\times n}$, $V \in \RR^{d \times d}$ &mdash; ортогональные матрицы
$$
U^TU = UU^T = I_n, \quad V^TV = VV^T = I_d,
$$
а $D \in \RR^{n \times d}$ &mdash; диагональная матрица. Тогда для оценки вектора весов по обучающей выборке $(
\boldX, \boldy)$ получаем:
$$
\hatboldw = (\boldX^T\boldX)^{-1}\boldX^T\boldy = (VD^TU^T UDV^T)^{-1}VD^TU^T\boldy = (VD^TDV^T)^{-1}VD^TU^T = V(D^TD)^{-1}V^T VD^TU^T = V(D^TD)^{-1}D^TU^T
$$

$$
D = \begin{pmatrix}
\lambda_1 &0   &0 &\dots &0\\
0   &\lambda_2 &0 &\dots &0\\
0   &0   &\lambda_3 &\dots &0\\
\vdots &\vdots &\vdots &\ddots &\vdots\\
0   &0   &0   &\dots &\lambda_d\\
0   &0   &0   &\dots &0\\
\vdots   &\vdots   &\vdots   &\dots &\vdots\\
0   &0   &0   &\dots &0
\end{pmatrix} = 
\begin{pmatrix}
\Lambda\\
\boldO_{(n-d)\times d}
\end{pmatrix},\quad D \in \RR^{n\times d},~\Lambda \in \RR^{d \times d},~ \boldO_{(n -d)\times d} \in \RR^{(n - d)\times d}
$$

$$
D^TD = \begin{pmatrix}
\Lambda &\boldO_{d\times (n - d)}
\end{pmatrix} \begin{pmatrix}
\Lambda\\
\boldO_{(n-d)\times d}
\end{pmatrix} = \Lambda^2 = \begin{pmatrix}
\lambda_1^2 &0 &\dots &0\\
0           &\lambda_2^2 &\dots &0\\
\vdots      &\vdots &\ddots &\vdots\\
0           &0      &0      &\lambda_d^2
\end{pmatrix} \Rightarrow (D^TD)^{-1} = \Lambda^{-2}
$$

$$
(D^TD)^{-1}D^T = \begin{pmatrix}
\Lambda^{-1} &\boldO_{d\times(n-d)}
\end{pmatrix} \in \RR^{d\times n}
$$

$$
\hatboldw = V \cdot \begin{pmatrix}
\Lambda^{-1} &\boldO_{d\times(n-d)}
\end{pmatrix} \cdot U^T
$$

$$
\frac{d\hatboldw}{dx_{ij}} = \frac{dV}{dx_{ij}} \cdot \begin{pmatrix}
\Lambda^{-1} &\boldO_{d\times(n-d)}
\end{pmatrix} \cdot U^T +
V \cdot \begin{pmatrix}
\frac{d \Lambda^{-1}}{d_{x_{ij}}} &\boldO_{d\times(n-d)}
\end{pmatrix} \cdot U^T+
V \cdot \begin{pmatrix}
\Lambda^{-1} &\boldO_{d\times(n-d)}
\end{pmatrix} \cdot \frac{dU^T}{d_{x_{ij}}}
$$

Пусть $\boldX = \hatU \Lambda \hatV^T$ &mdash; усеченное SVD-разложение. Тогда
$$
\frac{d\hatboldw}{dx_{ij}} = \frac{d \hatV}{d x_{ij}} \Lambda^{-1} \hatU^T + \hatV \frac{d\Lambda^{-1}}{dx_{ij}}\hatU^T + \hatV \Lambda^{-1}\frac{d\hatU^T}{dx_{ij}}, \quad \frac{d\Lambda^{-1}}{dx_{ij}} = \Lambda^{-2}\frac{d\Lambda}{dx_{ij}}
$$

$$
\hatboldw(\boldX, \boldy) = (\boldX^T\boldX)^{-1}\boldX^T\boldy
$$

$$
\hatboldw(\boldX + \Delta\boldX, \boldy + \Delta \boldy) = (\boldX^T\boldX)^{-1}\boldX^T\boldy + \Sum_{i,j} \frac{d\hatboldw}{d x_{ij}} \Delta x_{ij} + \Sum_{i} \frac{d\hatboldw}{d y_{i}} \Delta y_i + o(\Delta\boldX, \Delta \boldy),
$$
где $\Delta x_{ij} \sim \Normal(0, \sigma^2_{ij})$, $\Delta y_i \sim \Normal(0, \sigma^2)$

$$
\Cov(\hatboldw) \approx \Exp\ls \lp\Sum_{i,j} \frac{d\hatboldw}{d x_{ij}} \Delta x_{ij} + \Sum_{i} \frac{d\hatboldw}{d y_{i}} \Delta y_i \rp \lp \Sum_{i,j} \frac{d\hatboldw}{d x_{ij}} \Delta x_{ij} + \Sum_{i} \frac{d\hatboldw}{d y_{i}} \Delta y_i \rp^T\rs =
\Exp \ls \Sum_{i,j} \frac{d\hatboldw}{d x_{ij}} \lp \frac{d\hatboldw}{d x_{ij}}\rp^T\Delta x_{ij}^2 + \Sum_{i}
\frac{d\hatboldw}{d y_{i}}  \lp \frac{d\hatboldw}{d y_{i}}\rp^T\Delta y_i^2\rs = 
\Exp \ls \Sum_{i,j} \frac{d\hatboldw}{d x_{ij}} \lp \frac{d\hatboldw}{d x_{ij}}\rp^T\Delta x_{ij}^2 \rs  + \sigma^2 (\boldX^T\boldX)^{-1}.
$$

$$
\frac{d\hatboldw}{dx_{ij}} = \frac{d \hatV}{d x_{ij}} \Lambda^{-1} \hatU^T + \hatV \Lambda^{-2}\frac{d\Lambda}{dx_{ij}}\hatU^T + \hatV \Lambda^{-1}\frac{d\hatU^T}{dx_{ij}}
$$

$$
\frac{d\hatboldw}{dx_{ij}}  \lp \frac{d\hatboldw}{dx_{ij}} \rp^T = 
\lp  \frac{d \hatV}{d x_{ij}} \Lambda^{-1} \hatU^T + \hatV \Lambda^{-2}\frac{d\Lambda}{dx_{ij}}\hatU^T + \hatV \Lambda^{-1}\frac{d\hatU^T}{dx_{ij}}\rp
\lp \hatU  \Lambda^{-1} \frac{d \hatV^T}{d x_{ij}} + \hatU \frac{d\Lambda}{dx_{ij}}\Lambda^{-2}\hatV^T + \frac{d\hatU}{dx_{ij}} \Lambda^{-1}\hatV^T\rp = \\
= \frac{d \hatV}{d x_{ij}} \Lambda^{-2} \frac{d \hatV^T}{d x_{ij}} +
\frac{d \hatV}{d x_{ij}} \Lambda^{-3}  \frac{d\Lambda}{dx_{ij}} \hatV +
\frac{d \hatV}{d x_{ij}}\Lambda^{-1} \hatU^T\frac{d\hatU}{dx_{ij}}\Lambda^{-1}\hatV^T +
\hatV \Lambda^{-3}\frac{d\Lambda}{dx_{ij}}\frac{d\hatV^T}{dx_{ij}}+ \dots
$$

### Производная SVD-разложения<sup>[toc](#toc)</sup>

$$
\boldX = UDV^T \in \RR^{n \times m}
$$

\begin{gather*}
\frac{d\boldX}{dx_{ij}} = \frac{dU}{dx_{ij}}DV^T + U\frac{dD}{dx_{ij}}V^T + UD\frac{dV^T}{dx_{ij}}\\
U^TU = I_n \Rightarrow \frac{dU^T}{dx_{ij}} U + U^T \frac{dU}{dx_{ij}} = 0\\
V^TV = I_d \Rightarrow \frac{dV^T}{dx_{ij}} V + V^T \frac{dV}{dx_{ij}} = 0.
\end{gather*}

$$
\Omega_U^{ij} = U^T\frac{dU}{dx_{ij}} \in \RR^{n \times n},\quad \Omega_V^{ij} = \frac{dV^T}{dx_{ij}}V \in \RR^{m \times m}
$$

В данных обозначених получаем, что
$$
\lp\Omega_U^{ij}\rp^T + \Omega_U^{ij} = O,\\
\Omega_V^{ij} + \lp\Omega_V^{ij}\rp^T = O
$$

$$
U^T\frac{d\boldX}{dx_{ij}} V = \Omega_{U}^{ij} D + \frac{dD}{dx_{ij}} + D\Omega_V^{ij}
$$

* Матрицы $\Omega_U^{ij}$ и $\Omega_V^{ij}$ антисимметричны
* $d\boldX/dx_{ij} = E_{ij}$

$$
G \triangleq U^T\frac{d\boldX}{dx_{ij}} V = U^TE_{ij}V = \begin{pmatrix}
\boldu_1^T\\
\boldu_2^T\\
\vdots\\
\boldu_n^T
\end{pmatrix} 
\begin{pmatrix} 0 &0 &\dots &\bolde_i &\dots &0\end{pmatrix} \begin{pmatrix} \boldv_1 &\dots \boldv_d\end{pmatrix} =
\begin{pmatrix}
0 &0 &\dots &\boldu_1^T\bolde_i &\dots &0\\
0 &0 &\dots &\boldu_2^T\bolde_i &\dots &0\\
\dots &\dots &\dots &\dots &\dots &0\\
0      &0     &\dots &\boldu_n^T\bolde_i &\dots &0\\
\end{pmatrix}
\begin{pmatrix} \boldv_1 &\dots \boldv_d\end{pmatrix} =
\begin{pmatrix}
\boldu_1^T\bolde_i v_{j1} & \boldu_1^T\bolde_iv_{j2} &\dots\\
\boldu_2^T\bolde_i v_{j1} & \boldu_2^T\bolde_iv_{j2} &\dots\\
\vdots                   &\vdots                    &\ddots
\end{pmatrix} = 
\begin{pmatrix}
u_{i1} v_{j1} & u_{i1} v_{j2} &\dots\\
u_{i2} v_{j1} & u_{i2} v_{j2} &\dots\\
\vdots                   &\vdots                    &\ddots
\end{pmatrix} \in \RR^{n\times m}
$$
Здесь для удобства введено обзначение
$$
G \triangleq U^T\frac{d\boldX}{dx_{ij}} V
$$

Далее будем считать, что $n \ge m$. 

Рассмотрим $k$-ый диагональный элемент матрицы $G$ $(k \in \{1, \dots, m\})$. Выше определили, что
$$
\|G\|_{kk} = u_{ik}v_{jk}
$$
Таким образом, для диагональных элементов получаем
$$
\partfrac{d_k}{x_{ij}} = u_{ik}v_{jk}
$$
В результате полностью находим матрицу
$$
\partfrac{D}{x_{ij}} =
\begin{pmatrix}
\partfrac{d_1}{x_{ij}} &0 &\dots &0\\
0  &\partfrac{d_2}{x_{ij}} &\dots &0\\
\vdots &\vdots &\ddots &\vdots\\
0      &0      &\dots  &\partfrac{d_m}{x_{ij}}\\
0      &0      &\dots  &0\\
\vdots &\vdots &\dots  &\vdots\\
0      &0      &\dots  &0
\end{pmatrix} \in \RR^{n \times m}
$$

Рассмотрим недиагональный элемент $\|G\|_{kl}$

Далее рассмотрим элемент в позиции $kl$ ($k \in \{1, \dots, m\}$, $l \in \{1, \dots, m\}$:
$$
\|\Omega_U^{ij}\|_{kl} d_l + \| \Omega_V^{ij}\|_{kl}d_k = u_{ik}v_{jl} 
$$
Для транспонированного элемента получаем:
$$
\|\Omega_U^{ij}\|_{lk} d_k + \| \Omega_V^{ij}\|_{lk}d_l = u_{il}v_{jk}.
$$
Но из антисимметричности следует, что
$$
\|\Omega_U^{ij}\|_{kl} d_k + \| \Omega_V^{ij}\|_{kl}d_l = -u_{il}v_{jk},
$$
т.е. приходим к системе на значения элементов $\|\Omega_U^{ij}\|_{kl}$,  $\| \Omega_V^{ij}\|_{kl}$:
$$
\begin{cases}
&\|\Omega_U^{ij}\|_{kl} d_l + \| \Omega_V^{ij}\|_{kl}d_k = u_{ik}v_{jl},\\
&\|\Omega_U^{ij}\|_{kl} d_k + \| \Omega_V^{ij}\|_{kl}d_l = -u_{il}v_{jk}.
\end{cases}
$$

Теперь рассмотрим элемент $\|G\|_{kl}$, где $k \in \{m+1, \dots, n\}$, $l \in \{1, \dots, m\}$:
$$
u_{ik}v_{jl} = \|\Omega_U^{ij}\|_{kl} d_l,
$$
откуда восстанавливаем оставшиеся значения для $\Omega_U^{ij}$.

Теперь, когда  $\Omega_U^{ij}$ и  $\Omega_V^{ij}$ известны:
$$
\partfrac{U}{x_{ij}} = U\Omega_U^{ij},\\
\partfrac{V}{x_{ij}} = -V\Omega_V^{ij}.
$$

Сложность:
$$\lp \frac{M(M-1)}{2} + M + (N - M) * M \rp NM = O(N^3M)$$

### Обучение модели линейной регрессии<sup>[toc](#toc)</sup>

```python
def find_weights(X, y):
    """Finds the OLS/MLE weights estimate for multivariate linear regression model

    :param X: Matrix (np.ndarray) of shape (N, D) where N is number of objects and D is
        dimensionality of an object's feature vector
    :param y: Matrix of known dependent variables
    """
    assert X.ndim == 2 and y.ndim == 1
    assert X.shape[0] == y.shape[0]
    return np.linalg.inv(np.dot(X.T, X)).dot(X.T).dot(y)


class LinearRegression(object):
    def __init__(self, fit_intercept=True):
        super(LinearRegression, self).__init__()
        self._fit_intercept = fit_intercept
        self._weights = None
        
    def fit(self, X, y):
        X = np.array(X, dtype=np.float64, copy=False)
        N, D = X.shape
        assert N > D

        if self._fit_intercept:
            X = np.hstack([X, np.ones((N, 1))])
            D += 1

        assert X.shape == (N, D)
        assert y.shape == (N,)

        self._weights = find_weights(X, y)
        self._sigma2 = np.sum((np.dot(X, self._weights) - y)**2) / (N - D)
        self._covariance = self._sigma2 * np.linalg.inv(np.dot(X.T, X))
    
    def RSS(self, X, y):
        if self._fit_intercept:
            N = X.shape[0]
            X = np.hstack([X, np.ones((N, 1))])
        return np.sum((np.dot(X, self._weights) - y)**2)
    
    def MSE(self, X, y):
        return self.RSS(X, y) / X.shape[0]

    def predict(self, X):
        if self._fit_intercept:
            N = X.shape[0]
            X = np.hstack([X, np.ones((N, 1))])
        return np.dot(X, self._weights)
    
    @property
    def weights(self):
        return self._weights
    
    @property
    def sigma2(self):
        return self._sigma2
    
    @property
    def covariance(self):
        return self._covariance
    
    @property
    def variances(self):
        D = self._covariance.shape[0]
        return self._covariance[np.arange(D), np.arange(D)]
    
    @property
    def p_values(self):
        means = self.weights
        variances = self.variances
        p_values = 2 * scipy.stats.norm.cdf(-np.abs(means) / np.sqrt(variances))
        return p_values
```

# Регуляризация<sup>[toc](#toc)<sup>
* [Ridge-регрессия](#l2_regularization)
* [Примеры примения регуляризации](#regularization_examples)

Наиболее распространнеными являются следующие три типа регуляризации
* $\ell_1$-регуляризация. Предполагает нахождение весов путем минимизации следующей функции потерь:
$$
Q(\boldw;\boldX,\boldy) = \| \boldX\boldw - \boldy\|_2^2 + \lambda \|\boldw\|_1
$$
Алгоритм построения функции регрессии с использованием $\ell_1$-регуляризации также известен как **Lasso**.
* $\ell_2$-регуляризация. Предполагаем нахождение весов путем минимизации следующей функции потерь:
$$
Q(\boldw;\boldX,\boldy) = \| \boldX\boldw - \boldy\|_2^2 + \lambda \|\boldw\|_2^2
$$
Алгоритм построения функции регрессии с использованием $\ell_2$-регуляризации также известен как **ridge-регрессия** или **гребневая регрессия**.
* $\ell_p$-регуляризация
$$
Q(\boldw;\boldX,\boldy) = \| \boldX\boldw - \boldy\|_2^2 + \lambda \|\boldw\|_p^p
$$
* Elastic-Net
$$
Q(\boldw;\boldX,\boldy) = \| \boldX\boldw - \boldy\|_2^2 + \lambda \|\boldw\|_1 + \tau \|\boldw\|_2^2
$$

<a id='l2_regularization'></a>
## Ridge-регрессия ($\ell_2$-регуляризация)<sup>[toc](#toc)</sup>

### Аналитическое выражение для вектора весов<sup>[toc](#toc)</sup>

$$
Q(\boldw;\boldX,\boldy) = \| \boldX\boldw - \boldy\|_2^2 + \lambda \|\boldw\|_2^2 \to \min_{\boldw}
$$

$$
\frac{dQ(\boldw;\boldX,\boldy)}{d\boldw} = \frac{d}{d\boldw}\lf (\boldX\boldw - \boldy)^T(\boldX\boldw - \boldy) + \lambda \|\boldw\|_2^2 \rf = 2\boldX^T\boldX\boldw - 2\boldX^T\boldy + 2\lambda\boldw = \boldzero
$$

$$
\hatboldw = (\boldX^T\boldX + \lambda \boldI)^{-1}\boldy
$$

### Вероятностная интерпретация $\ell_2$-регуляризации<sup>[toc](#toc)</sup>

Введем априорное распределение $p(\boldw)$ на вектор весов $\boldw$. Пусть $\boldw \sim \Normal(\boldzero, \alpha\boldI)$. Тогда для правдоподобия выборки получим:
$$
\mathcal{L}(\boldX,\boldy;\boldw,\sigma^2) = \log \lp P(\boldX,\boldy|\boldw, \sigma^2) p(\boldw) \rp = 
\text{const} - \frac{n}{2}\log \sigma^2 - \frac{1}{2\sigma^2} \|\boldX\boldw - \boldy\|_2^2 - \frac{1}{2\alpha}\|\boldw\|_2^2 \to \max_{\boldw, \sigma^2}
$$

Максимизация правдоподобия по $\boldw$ эквивалентна минимизации следующего функционала
$$
Q(\boldw;\boldX,\boldy) = \frac{1}{\sigma^2} \|\boldX\boldw - \boldy\|_2^2 + \frac{1}{\alpha}\|\boldw\|_2^2,
$$
т.е. задаче регрессии с $\ell_2$-регуляризацией, где
$$
\lambda = \frac{\sigma^2}{\alpha}.
$$

Если шум $\sigma^2$ неизвестен, то можно находить $\boldw$ и $\sigma^2$ итеративно:
\begin{align}
&\hatboldw^{(0)} = (\boldX^T\boldX)^{-1}\boldX\boldy\\
&{\sigma^2}^{(0)} = \frac{1}{n}\|\boldX\hatboldw^{(0)}  - \boldy\|_2^2
\end{align}
\begin{align}
&\hatboldw^{(i)} = \lp\boldX^T\boldX+ \frac{{\sigma^2}^{(i-1)}}{\alpha}\boldI\rp^{-1}\boldy\\
&{\sigma^2}^{(i)} = \frac{1}{n}\|\boldX\hatboldw^{(i-1)}  - \boldy\|_2^2
\end{align}

```python
def get_model_path(make_model, alphas, **model_kwargs):
    coefs = []
    for a in alphas:
        model = make_model(alpha=a, **model_kwargs)
        model.fit(X, y)
        coefs.append(model.coef_)
    return np.array(coefs)
```

```python
def plot_model_path(ax, alphas, Model, **kwargs):
    model_name = str(Model())
    model_name = model_name[:model_name.find('(')]

    coeffs = get_model_path(Model, alphas, **kwargs)

    ax.plot(alphas, coeffs)
    ax.set_xscale('log')
    ax.set_xlabel('regularization parameter')
    ax.set_ylabel('weights')
    ax.set_title('{} coefficients'.format(model_name))
    ax.grid(which='major', linestyle='--', alpha=0.5)
```

# RANSAC<sup>[toc](#toc)</sup>

<a id='ransac_algorithm'></a>
## Алгоритм<sup>[toc](#toc)</sup>
* Входные данные: обучающая выборка $(\boldX, \boldy)$, $\boldX \in \RR^{n\times d}$, $\boldy \in \RR^{n}$
* Параметры алгоритма:
    * $N_{samples}$ размер подвыборок
    * $T_{outlier}$ некоторый порог, по которому будут определяться аутлайер

**Алгортим:**
* Инициализируем текущие значения:
    * Максимальное достигнутое количество инлайеров $\hatN_{inliers} = 0$.
    * Минимальная метрика качества ($RSS$).
    * Вектор весов $\hatboldw$ на котором достигнуты значения выше, считаем равным `None`.
* Выбираем подмножество из $N_{samples}$
    * Обучаем модель на подмножестве из выбранных $N_{samples}$ объектов
    * Опеределяем отклонения (residuals) фактического значения зависимой переменной от предсказанного значения для всех элементов выборки
    * Делим объекты на инлаеры и аутлайеры по порогу $T_{outlier}$
    * Если количество инлайеров меньше, чем уже достигнутое $\hatN_{inliers}$, то переходим к следующей итерации
    * Если количество инлайеров больше, чем уже достигнутое, то текущие параметры модели считаем лучшими
    * Если количество инлайеров равно уже достигнутому $\hatN_{inliers}$, то сравниваем модели по метрике качества $RSS$ и выбираем лучшую

<a id='ransac_examples'></a>
## Примеры использования<sup>[toc](#toc)</sup>

```python
np.random.seed(1)

n_samples = 1000
n_outliers = 50

X, y, coef = datasets.make_regression(
    n_samples=n_samples, n_features=1,
    n_informative=1, noise=10,
    coef=True, random_state=0)

assert X.shape == (n_samples, 1)
assert y.shape == (n_samples,)
print('Real coefficients: {}'.format(coef))

# Adding outliers
X[:n_outliers] = 3 + 0.5 * np.random.normal(size=(n_outliers, 1))
y[:n_outliers] = -3 + 10 * np.random.normal(size=n_outliers)

# Fit line using all data
lr = linear_model.LinearRegression()
lr.fit(X, y)
print('Linear regression coefficients: {}'.format(lr.coef_))

# Robustly fit linear model with RANSAC algorithm
ransac = linear_model.RANSACRegressor(residual_threshold=20)
ransac.fit(X, y)
inlier_mask = ransac.inlier_mask_
outlier_mask = np.logical_not(inlier_mask)
print('RANSAC regression coefficients: {}'.format(ransac.estimator_.coef_))

# Predict data of estimated models
line_X = np.arange(X.min(), X.max())[:, np.newaxis]
line_y = lr.predict(line_X)
line_y_ransac = ransac.predict(line_X)

lw = 2
plt.figure(figsize=(10, 10))
plt.scatter(X[inlier_mask], y[inlier_mask], color='blue', marker='.', alpha=0.7, label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask], color='red', marker='.',label='Outliers')
plt.plot(line_X, line_y, color='red', linewidth=lw, label='Linear regressor')
plt.plot(line_X, line_y_ransac, color='blue', linewidth=lw, label='RANSAC regressor')
plt.legend(loc='lower right')
plt.xlabel("$x$");
plt.ylabel("$y$");
```

# Байесовская регрессия<sup>[toc](#toc)<sup>

В $\ell_2$-регуляризации мы находили максимум правдоподобия при условии, что $p(\boldw) \sim \Normal(\boldzero, \alpha\boldI)$. Теперь же попробуем найти апостериорное распределение на вектор весов $\boldw$ при условии выборки $(\boldX,\boldy)$:

$$
p(\boldw|\boldX,\boldy) = \frac{p(\boldw,\boldX,\boldy)}{p(\boldX,\boldy)} = \frac{p(\boldy|\boldX,\boldw)p(\boldX,\boldw)}{p(\boldX,\boldy)} = \frac{p(\boldy|\boldX,\boldw)p(\boldw)p(\boldX)}{p(\boldX,\boldy)} = \frac{p(\boldy|\boldX,\boldw)p(\boldw)}{p(\boldy|\boldX)} = \frac{p(\boldy|\boldX,\boldw)p(\boldw)}{\Int p(\boldy|\boldX, \boldw')p(\boldw')d\boldw'} \propto p(\boldy|\boldX,\boldy)p(\boldw)
$$

Раньше находили только максимум величины справа, т.е. моду распределения $p(\boldw|\boldX,\boldy)$. Теперь же найдем само распределение.
$$
p(\boldy|\boldX,\boldy)p(\boldw) =  C\cdot \exp\lf- \frac{1}{2\sigma^2}\|\boldX\boldw - \boldy\|_2^2 - \frac{1}{2\alpha}\|\boldw\|_2^2 \rf
$$

Соберем полный квадрат под экспонентой:
$$
\frac{1}{\sigma^2}(\boldw^T\boldX^T - \boldy^T)(\boldX\boldw - \boldy) + \frac{1}{\alpha}\boldw^T\boldw = 
\boldw^T\boldX^T\boldX\boldw - \frac{1}{\sigma^2}\boldw^T\boldX^T\boldy - \frac{1}{\sigma^2}\boldy^T\boldX\boldw + \frac{1}{\alpha}\boldw^T\boldw + \frac{1}{\sigma^2}\boldy^T\boldy = 
\boldw^T(\sigma^{-2}\boldX^T\boldX + \alpha^{-1})\boldw - \boldw^T\sigma^{-2}\boldX^T\boldy - \boldy^T\sigma^{-2}\boldX\boldw + \text{const}
$$

$$
(\boldw - \boldm)^T\Sigma^{-1}(\boldw -\boldm) = \boldw^T\Sigma^{-1}\boldw - \boldm^T\Sigma^{-1}\boldw -\boldw^T\Sigma^{-1}\boldm + \boldm^T\Sigma^{-1}\boldm.
$$

Отсюда следует, что
$$
\Sigma^{-1} = \sigma^{-2}\boldX^T\boldX + \alpha^{-1} = \sigma^{-2} \lp \boldX^T\boldX + \frac{\sigma^2}{\alpha}I_d\rp\\
\Sigma = \sigma^2 \lp \boldX^T\boldX + \frac{\sigma^2}{\alpha}I_d\rp^{-1},\\
\boldm = \Sigma \cdot \sigma^{-2}\boldX^T\boldy =  \lp \boldX^T\boldX + \frac{\sigma^2}{\alpha}I_d\rp^{-1}\boldX^T\boldy
$$
$\boldm$ является модой апостериорного распределения. Эту моду мы уже нашли при решении задачи $\ell_2$-регуляризации.

$$
p(\boldw|\boldX,\boldy) = \Normal\lp\boldw|\lp \boldX^T\boldX + \frac{\sigma^2}{\alpha}I_d \rp^{-1}\boldX^T\boldy, \sigma^{-2} \lp \boldX^T\boldX + \frac{\sigma^2}{\alpha}I_d\rp^{-1}\rp
$$

* Нулевая априорная информации о $\boldw$ математически соответствует бесконечной априорной дисперсии $\alpha \to \infty$. В таком случае получаем
$$
\boldm = (\boldX^T\boldX)^{-1}\boldX^T\boldy,
$$
т.е. результат для линейной регрессии.
* При нулевой априорной дисперсии, т.е. когда мы владеем абсолютным знанием о $\boldw$ и $\alpha \to 0$, получаем, что
$$
\boldm \to \boldzero.
$$

# Выбор модели/отбор признаков<sup>[toc](#toc)</sup>

* $(\boldX, \boldy)$ &mdash; обучающая выборка: $\boldX \in \RR^{n\times d}$, $\boldy \in \RR^n$
* $a(\cdot; \boldX, \boldy)$ &mdash; некоторым алгоритм предсказания, построенный по обучающей выборке $(\boldX,\boldy)$

Пусть известно совместное распределение $F_{\boldx,y}$.
* $r(\boldx) = \Exp_{y|\boldx} (y) = \Int yp(y|\boldx)dy$ &mdash; функция регрессии
* $a(\cdot) \triangleq a(\cdot;\boldX,\boldy)$ &mdash; аппроксимация функции регрессии

$$
\Risk(a) = \Exp_{\boldX,\boldy} \ls \Exp_{\boldx,y} (y - a(x;\boldX,\boldy))^2 \rs
$$

$$
\bar{a}(x) = \Exp_{\boldX,\boldy} \ls a(\boldx;\boldX,\boldy) \rs
$$

Можно показать, что
$$
\Risk(a) = \underbrace{\Exp_{\boldx,y} (y - r(\boldx))^2}_{шум (noise)} +
\underbrace{\Exp_{\boldx}(r(\boldx) - \bar{a}(\boldx))^2}_{смещение (bias)} +
\underbrace{\Exp_{\boldX,\boldy}\Exp_{\boldx}(a(\boldx;\boldX,\boldy) - \bar{a}(\boldx))^2}_{вариация (variance)},
$$
т.е. **bias variance decomposition**.
 
На практике не знаем $F_{\boldx,y}$, поэтому нужно уметь оценивать все по самой выборке $(\boldX,\boldy)$ 

$\newcommand{\tr}{\mathrm{tr}}$

Давайте предположим, что пока что знаем условные распределения $F_{y|\boldx}$, но не знаем $F_{\boldx}$. Вместо распределения $F_{\boldx}$ фактически возьмем э.ф.р. $\hatF_{\boldx}$.
$$
\Risk(a;\boldX) = \frac{1}{n}\Exp_{\boldy|\boldX} \ls \Sum_{i = 1}^n \Exp_{y^*|\boldx_i}\lp y^* - a(\boldx_i; \boldX, \boldy)\rp^2 \rs
$$

Теперь частично откажемся от знания распределения $F_{y|\boldx}$, фактически заменяя его в каждой точке на э.ф.р. $\hatF_{y|\boldx}$:
$$
\Risk_{\tr}(a;\boldX,\boldy) = \frac{1}{n} \Exp_{\boldy|\boldX} \ls \Sum_{i = 1}^n \lp y_i - a(\boldx_i; \boldX, \boldy)\rp^2 \rs
$$

* $\Risk(a;\boldX)$ &mdash; **ошибка на предсказании**
* $\Risk_{\tr}(a;\boldX,\boldy)$ &mdash; **ошибка на обучении**

Естественно ожидать, что
$$\Risk_{\tr}(a;\boldX,\boldy) \le \Risk(a;\boldX)$$
Можно показать, что
$$
\Risk(a;\boldX) = \Risk_{\tr}(a;\boldX,\boldy) + \frac{2}{n} \Sum_{i = 1}^n \Cov_{\boldy|\boldX}(a(\boldx_i), y_i)
$$

### Статистика Mallow $C_p$
В случае модели линейной регрессии
$$
\boldy = \boldX\boldw + \boldeps, \quad \boldX \in \RR^{n\times d}, \quad \boldy \in \RR^{n}, \quad \Exp\boldeps = \boldzero, \quad \Var \boldeps = \sigma^2 \boldI
$$
можно показать, что
$$
\Risk(a;\boldX) = \Risk_{\tr}(a;\boldX,\boldy) + \frac{2d\sigma^2}{n}.
$$

Но распределение $F_{y|\boldx}$ нам все также неизвестно, поэтому требуется отказтаься и от него
\begin{align}
&\hat{\Risk}_{\tr}(a;\boldX,\boldy) = \frac{1}{n}\Sum_{i=1}^n(a(\boldx_i;\boldX,\boldy) - y_i)^2 = \frac{1}{n}\|\boldX\hatboldw - \boldy\|^2,\\
&\hat{\Risk}(a;\boldX,\boldy) = \hat{\Risk}_{\tr}(a;\boldX,\boldy) + \frac{2d\hatsigma^2}{n},\\
&\hatsigma^2 = \frac{1}{n - d}\Sum_{i = 1}^n(y_i - a(\boldx_i;\boldX,\boldy)^2)
\end{align}
Поэтому
$$
\hat{\Risk}(a;\boldX,\boldy) = \frac{1}{n}\|\boldX\hatboldw - \boldy\|^2 + \frac{2d}{n} \frac{1}{n - d} \|\boldX\hatboldw - \boldy\|_2^2 =\frac{n + d}{n - d} \cdot \frac{1}{n} \|\boldX\hatboldw - \boldy\|^2
$$

### LOO-проверка<sup>[toc](#toc)</sup>
$$
\hat{\Risk}(a;\boldX,\boldy) = \frac{1}{n}\Sum_{i = 1}^n (a(\boldx_i, \boldX^{-i}, \boldy^{-i}))
$$

$$
\hat{\Risk}(a;\boldX,\boldy) = \Sum_{i = 1}^n \lp \frac{y_i - a(x_i;\boldX,\boldy)}{1 - \|\boldX(\boldX^T\boldX)^{-1}\boldX^T\|_{ii}}\rp^2

### Критерии<sup>[toc](#toc)</sup>

Рассмотрим несколько критериев для оценивания модели.

```python
class Criterion(object):
    def initialize(self, X, y):
        self._check_Xy_pair(X, y)
        self.X = X; self.y = y
        n_samples, n_features = X.shape
        self.weights = find_weights(X, y)
        self.sigma2 = np.sum((y - np.dot(X, self.weights))**2) / (n_samples - n_features)
        self.sigma = np.sqrt(self.sigma2)
        return self

    def _check_Xy_pair(self, X, y):
        assert isinstance(X, np.ndarray)
        assert isinstance(y, np.ndarray)
        assert X.ndim == 2
        assert y.ndim == 1
        assert X.shape[0] == y.shape[0]

    def __call__(self, indices):
        assert False, "Not implmented"

    def __repr__(self):
        return 'Criterion'

class CriterionCp(Criterion):
    def __call__(self, indices):
        X = self.X[:, indices]
        n_samples, n_features = X.shape
        weights = find_weights(X, self.y)
        return (np.sum((np.dot(X, weights) - self.y)**2) + 2 * n_features * self.sigma2) / n_samples
    def __repr__(self):
        return 'Cp'

class CriterionLOO(Criterion):
    def __call__(self, indices):
        X = self.X[:, indices]
        weights = find_weights(X, self.y)
        n_samples, n_features = X.shape
        U = np.dot(np.dot(X, np.linalg.inv(np.dot(X.T, X))), X.T)
        U = U[np.arange(n_samples), np.arange(n_samples)]
        return np.sum(((self.y - np.dot(X, weights)) / (1 - U)) ** 2) / n_samples
    def __repr__(self):
        return 'LOO'

class CriterionCV(Criterion):
    def __init__(self, n_folds=5, random_state=1):
        self.n_folds = n_folds
        self.random_state = random_state
    def __call__(self, indices):
        X = self.X[:, indices]
        n_samples, n_features = X.shape
        kfold = KFold(n_splits=self.n_folds, shuffle=True, random_state=self.random_state)
        R_CV = 0
        for tr_indices, ts_indices in kfold.split(X):
            X_tr = X[tr_indices]; y_tr = self.y[tr_indices]
            X_ts = X[ts_indices]; y_ts = self.y[ts_indices]
            weights = find_weights(X_tr, y_tr)
            R_CV += np.sum((y_ts - np.dot(X_ts, weights)) ** 2)
        return R_CV / n_samples
    def __repr__(self):
        return 'CV' + str(self.n_folds)
    
class CriterionTest(Criterion):
    def __init__(self, X_test, y_test):
        self.X_test = np.array(X_test)
        self.y_test = np.array(y_test)
    def __call__(self, indices):
        X = self.X[:, indices]
        n_samples, n_features = X.shape
        weights = find_weights(X, self.y)
        return np.sum((self.y_test - np.dot(self.X_test[:, indices], weights))**2) / self.X_test.shape[0]
    def __repr__(self):
        return 'Test'
```

### Классы FullSearch и GreedySearch [[toc]](#toc) [[up]](#model_choice)

```python
class FullSearch:
    def __init__(self, X, y, criterion, X_test=None, y_test=None, feature_names=None, verbose=False):
        criterion.initialize(X, y)
        self.criterion = criterion
        self.X = X; self.y = y
        self.n_samples, self.n_features = X.shape
        if feature_names is None:
            feature_names = list(range(self.n_features))
        assert len(feature_names) == self.n_features
        self.feature_names = feature_names
        self.fname2ind = {name:i for i, name in enumerate(self.feature_names)}
        self.ind2fname = {i:name for i, name in enumerate(self.feature_names)}

        if not ((X_test is None) | (y_test is None)):
            self.criterion._check_Xy_pair(X_test, y_test)
            assert self.X.shape[1] == self.n_features
            self.X_test = X_test
            self.y_test = y_test
        else:
            self.X_test = None
            self.y_test = None

    def find_best_subset(self):
        best_value = np.inf
        best_subset = []
        for n_features in range(1, self.n_features + 1):
            for indices in combinations(list(range(self.n_features)), n_features):
                criterion_value = self.criterion(indices)
                if criterion_value < best_value:
                    best_value = criterion_value
                    best_subset = indices
        feature_names = [self.ind2fname[i] for i in best_subset]
        return OrderedDict([('feature_indices', best_subset), 
               ('feature_names', feature_names), 
               ('best_value', best_value)])

class GreedySearch:
    def __init__(self, X, y, criterion, feature_names=None, forward=True, verbose=False):
        self.X = np.array(X)
        self.y = np.array(y)
        self.criterion = criterion.initialize(self.X, self.y)
        self.n_samles, self.n_features = self.X.shape
        if feature_names is None:
            feature_names = list(range(self.n_features))
        assert len(feature_names) == self.n_features
        self.feature_names = feature_names
        self.forward = forward
        self.verbose = verbose
        self.n_samples, self.n_features = X.shape
        self.all_features = set(range(self.n_features))
        self._find_sigma()
    
    def _find_sigma(self):
        self.beta = find_weights(self.X, self.y)
        self.sigma = np.sqrt(np.sum((self.y - np.dot(self.X, self.beta)) ** 2) / (self.n_samples - self.n_features))
           
    def __call__(self):
        if self.forward:
            self.curr_value = np.inf
            self.curr_features = set()
            while self._forward_step():
                pass
        else:
            self.curr_value = np.inf
            self.curr_features = set(range(self.n_features))
            while self._backward_step():
                pass
            
        best_subset = sorted(list(self.curr_features))
        feature_names = [self.feature_names[i] for i in best_subset]
        return OrderedDict([('feature_indices', best_subset), 
                            ('feature_names', feature_names), 
                            ('best_value', self.curr_value)])
            
    def _forward_step(self):
        self.not_considered = self.all_features - self.curr_features # Еще не рассморенные признаки
        best_feature = -1     
        best_value = np.inf
        for n_feature in sorted(list(self.not_considered)):
            new_features = copy.deepcopy(self.curr_features)
            new_features.add(n_feature)
            new_features = np.array(sorted(list(new_features)))

            value = self.criterion(new_features)
            self.print('F_step: new feature set {} gives value = {}'.format(new_features, value))
            if value < best_value:
                best_value = value
                best_feature = n_feature
        if best_value < self.curr_value:
            self.curr_value = best_value
            self.print('F_step: feature {} added to feature set {}\n'.format(
                best_feature, sorted(list(self.curr_features))))
            self.curr_features.add(best_feature)
            return True
        else:
            self.print('F_step: maximum reached for feature set {}; current value = {}\n'.format(
                self.curr_features, self.curr_value))
            return False
    
    def _backward_step(self):
        best_feature = -1
        best_value = np.inf
        for n_feature in sorted(list(self.curr_features)):
            new_features = copy.deepcopy(self.curr_features)
            new_features.remove(n_feature)
            new_features = np.array(sorted(list(new_features)))
            if len(new_features) == 0:
                break
            value = self.criterion(new_features)
            self.print('B_step: new feature set {} gives value = {}'.format(new_features, value))
            if value < best_value:
                best_value = value
                best_feature = n_feature
        if best_value < self.curr_value:
            self.curr_value = best_value
            self.print('B_step: feature {} removed from feature set {}\n'.format(
                best_feature, sorted(list(self.curr_features))))
            self.curr_features.remove(best_feature)
            return True
        else:
            self.print('B_step: maximum reached for feature set {}; current value = {}\n'.format(
                self.curr_features, self.curr_value))
            return False
        
    def print(self, msg):
        if self.verbose:
            print(msg)
```
